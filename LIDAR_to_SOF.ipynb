```python
import numpy as np
import pandas as pd
import  datetime 

from scipy.io import loadmat
```

```python
# Change this variable to read in another file as long as it's the same path
filename = 'Summary2018_Q1.mat'

base = './input/'
base_out = './output/'

output_file = filename.split('.')[0] + '.sof'
print('Will output results to', output_file)
```

```python
data = loadmat(base+filename)
```

```python
# for key in data.keys():
#     print(key)
# #data
# #df = pd.DataFrame(data=data, index=[0])
```

```python
# _direc is direction
# _doy is day of year
# _range is height
# _sdate is datenum in matlab (julian day since epoch)
# _site is name of site
# _u is wind speed

wanted_keys = ['out_direc_h1', 'out_doy_h1', 'out_sdate_h1', 'out_u_h1'] # The keys you want

d = {}
for key in wanted_keys:
    d[key]=data[key].T[0]  #This is needed for specific struct format
```

```python
df = pd.DataFrame(d)    
```

```python
df.dropna(how='all', inplace=True)
```

```python
df['date'] = df['out_sdate_h1'].map(lambda x: datetime.datetime(1,1,1) + datetime.timedelta(days=x) - datetime.timedelta(days=366))
```

```python
df.rename(columns={'out_direc_h1': 'Wind Direction' , 'out_u_h1':'Wind Speed' }, inplace=True)
```

```python
with open(base_out + output_file, 'w') as f:
    f.write(output_header + '\n')
    for i, row in df.iterrows():
        date = row['date'].timetuple()
        nextline = ','.join(['100' , str(date.tm_year),  str(date.tm_yday), '{:02d}{:02d}'.format(date.tm_hour,date.tm_min), str(date.tm_sec), str(row['Wind Speed']),str(row['Wind Direction'] )])
        f.write(nextline + '\n')
 
    
```
